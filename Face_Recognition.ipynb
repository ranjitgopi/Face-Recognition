2. Write function to create metadata of the image.  
Hint: Metadata means derived information from the available data which can be useful for particular problem statement

```python
from PIL import Image
from PIL.ExifTags import TAGS

def imagemetadata(imagepath):
  
    image = Image.open(imagepath)
    exifdata= image.getexif()
    
    if exifdata == {}:
        print('No metadata')

    for tagid in exifdata:
        tagname = TAGS.get(tagid, tagid)
        value = exifdata.get(tagid)
        print(f"{tagname:25}: {value}")
```

4. Generate Embeddings vectors on the each face in the dataset.  
Hint: Use ‘vgg_face_weights.h5’

```python
from tensorflow.keras.layers import ZeroPadding2D, Convolution2D, MaxPooling2D, Dropout, Flatten, Activation
from tensorflow.keras.models import Sequential, Model
from sklearn.svm import SVC

model = Sequential()
model.add(ZeroPadding2D((1, 1), input_shape = (224, 224, 3)))
model.add(Convolution2D(64, (3, 3), activation = 'relu'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(64, (3, 3), activation = 'relu'))
model.add(MaxPooling2D((2, 2), strides = (2, 2)))

model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(128, (3, 3), activation = 'relu'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(128, (3, 3), activation = 'relu'))
model.add(MaxPooling2D((2, 2), strides = (2, 2)))

model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(256, (3, 3), activation = 'relu'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(256, (3, 3), activation = 'relu'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(256, (3, 3), activation = 'relu'))
model.add(MaxPooling2D((2, 2), strides = (2, 2)))

model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(512, (3, 3), activation = 'relu'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(512, (3, 3), activation = 'relu'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(512, (3, 3), activation = 'relu'))
model.add(MaxPooling2D((2, 2), strides =(2, 2)))

model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(512, (3, 3), activation = 'relu'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(512, (3, 3), activation = 'relu'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(512, (3, 3), activation = 'relu'))
model.add(MaxPooling2D((2, 2), strides=(2, 2)))

model.add(Convolution2D(4096, (7, 7), activation = 'relu'))
model.add(Dropout(0.5))
model.add(Convolution2D(4096, (1, 1), activation = 'relu'))
model.add(Dropout(0.5))
model.add(Convolution2D(2622, (1, 1)))
model.add(Flatten())
model.add(Activation('softmax'))
```

```python
model.load_weights('vgg_face_weights.h5')
print(model.summary())
```

```python
vgg_face_descriptor = Model(inputs = model.layers[0].input, outputs = model.layers[-2].output)
```

```python
import cv2
import os,glob
import matplotlib.pylab as plt
from os import listdir,makedirs
from os.path import isfile,join
import pandas as pd
import numpy as np

path=os.getcwd()+r'\PINS'

folders =  os.listdir(path)

data = []
X=[]
y=[]

for folder in folders:
    
    path_updated = path + '\\'+folder 

    files = os.listdir(path_updated)

    if files !=[]:

        #for file in files:
        for index in range(0,20):

            file=files[index]

            print('Processing image '+os.path.join(path_updated,file))
            img = cv2.imread(os.path.join(path_updated,file),cv2.IMREAD_COLOR)
            img = (img / 255.).astype(np.float32)
            img = cv2.resize(img, dsize = (224, 224))
            embedding_vector = vgg_face_descriptor.predict(np.expand_dims(img, axis = 0))[0]

            data.append([file[:-4],folder[5:],embedding_vector])
            
                    
            X.append(embedding_vector)
            y.append(folder[5:])
```

```python
df = pd.DataFrame(data, columns=['NameOfImage', 'Actor','Embedding Vector'])
```

```python
print('The data frame of actor images and embedded vectors:')
print(df)
```

5. Build distance metrics for identifying the distance between two similar and dissimilar images. 

```python
def distance(embeddedvector1, embeddedvector2):
    return np.sum(np.square(embeddedvector1 - embeddedvector2))
```

```python
def similarity(dataframeindex1,dataframeindex2):
        print('Checking similarity between faces of '+df['Actor'][dataframeindex1]+' and '+df['Actor'][dataframeindex2])
        similarityindex=distance(df['Embedding Vector'][dataframeindex1],df['Embedding Vector'][dataframeindex2])
        print('Similarity index: ',similarityindex)
```

7. Build an SVM classifier in order to map each image to its right person.  

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)

print("{0:0.2f}% data is in training set".format((len(X_train)/len(X)) * 100))
print("{0:0.2f}% data is in test set".format((len(X_test)/len(X)) * 100))
```

```python
X_train=np.array(X_train)
print('X_train shape:')
print(X_train.shape)
print(' ')

X_test=np.array(X_test)
print('X_test shape:')
print(X_test.shape)
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
Scaler = StandardScaler()

X_train = Scaler.fit_transform(X_train)
X_test = Scaler.transform(X_test)
```

```python
#Perform PCA
from sklearn.decomposition import PCA

n_components_used=100

pca = PCA(n_components=n_components_used)
pca.fit(X_train)

print("Eigen values of PCA")
print(pca.explained_variance_)
print(" ")

print("Eigen vectors of PCA")
print(pca.components_)
print(" ")

percentage_variance_by_eigen_vals=(pca.explained_variance_ratio_)*100
print("Percentage of variance explained by eigen vectors")
print(percentage_variance_by_eigen_vals)
print(" ")

plt.bar(list(range(1,n_components_used+1)),percentage_variance_by_eigen_vals,alpha=0.15, align='center')
plt.xticks(np.arange(0, n_components_used+1, 1.0))
plt.yticks(np.arange(0, 110, 10))
plt.ylabel('Variation explained by Eigen Value (%)')
plt.xlabel('Eigen Value')
plt.show()

plt.step(list(range(1,n_components_used+1)),np.cumsum(percentage_variance_by_eigen_vals), where='mid')
plt.axhline(y = 90, color = 'r', linestyle = '-')
plt.xticks(np.arange(0, n_components_used+1, 1.0))
plt.yticks(np.arange(0, 110, 10))
plt.ylabel('Cumulative of variation explained')
plt.xlabel('Eigen Value')
plt.show()

print(' ')
print('We will use 100 components in PCA for model reduction')
```

```python
pca = PCA(n_components=100)
pca.fit(X_train)

X_train= pca.transform(X_train)
X_test= pca.transform(X_test)
```

```python
y_train=np.array(y_train)
y_test=np.array(y_test)
```

```python
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)
```

```python
print("Labels:")
list(label_encoder.classes_)
```

```python
from sklearn.svm import SVC
model= SVC(C = 1, gamma = 0.001, kernel = 'rbf', class_weight = 'balanced', random_state = 100)
model.fit(X_train, y_train)
```

```python
print('SVC accuracy on train set:')
print(model.score(X_train, y_train))
```
